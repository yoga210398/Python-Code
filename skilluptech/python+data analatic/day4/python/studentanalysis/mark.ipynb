```python
import pandas as pd
# Read file where each CSV row is wrapped in quotes
raw = pd.read_csv('students.csv', header=None, dtype=str)
# remove outer quotes from each line if present
lines = raw.iloc[:, 0].astype(str).str.replace('^"|"$','', regex=True)
# split into columns and set header from first row
df = lines.str.split(',', expand=True)
df.columns = df.iloc[0]
df = df[1:].reset_index(drop=True)
# convert numeric columns
df['Marks'] = pd.to_numeric(df['Marks'], errors='coerce')
df['ID'] = pd.to_numeric(df['ID'], errors='coerce')
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Attendance'] = pd.to_numeric(df['Attendance'], errors='coerce')
df
```
